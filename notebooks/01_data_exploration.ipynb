```python
import pandas as pd
import os
import pathlib
from pathlib import Path
```

```python

data_directory = Path.cwd().parent / "data"

# find all CSV files under the data directory (returns Path objects)
CSV_FILE_PATHS = list(data_directory.glob('*.csv'))
print(f"Found {len(CSV_FILE_PATHS)} CSV files in {data_directory}")
```

```python
# Load the dataset
data_Frames = []

for file_path in CSV_FILE_PATHS:
    print(f"Loading file: {file_path.name}")
    # pandas accepts Path objects directly in recent versions
    df = pd.read_csv(file_path, low_memory=False)
    data_Frames.append(df)

# Combine all dataframes into a single dataframe
combined_df = pd.concat(data_Frames, ignore_index=True)
print(f"Combined DataFrame shape: {combined_df.shape}")
print(combined_df.head())

combined_df.to_csv(data_directory / "combined_dataset.csv", index=False)
# Display basic information about the dataset
```
